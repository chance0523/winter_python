### 데이터 프레임 생성

```python
import pandas as pd
```

```python
dict1 = {
    '2015' : [9904312, 3448737, 2890451, 2466052],
    '2010' : [9631434, 4134153, 1351353, 3256362],
    '2005' : [8249424, 1341153, 5462523, 5151353],
    '2000' : [9158513, 1513326, 3563623, 4632352],
    '지역' : ['수도권','경상권','수도권','경상권'],
    '증가율' : [0.0234, 0.0384, 0.3423, 0.3423]
}

option1=['지역','2000','2005','2010','2015','증가율']
option2 = ['서울','부산','인천','대구']

df1 = pd.DataFrame(dict1,columns=option1,index=option2)
df1
```

```python
list1 = [
    [9904312, 3448737, 2890451, 2466052],
    [9631434, 4134153, 1351353, 3256362],
    [8249424, 1341153, 5462523, 5151353],
    [9158513, 1513326, 3563623, 4632352],
    ['수도권','경상권','수도권','경상권'],
    [0.0234, 0.0384, 0.3423, 0.3423]
]

df2 = pd.DataFrame(list1)
df2
```
