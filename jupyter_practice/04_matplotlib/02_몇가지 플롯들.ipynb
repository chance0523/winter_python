### 바 차트

막대 그래프

bar - 수직 그래프

barh - 수평 그래프

```python
x = [1,2,3]
y = [2,3,1]

plt.title('기본 bar 차트')
plt.bar(x,y)

plt.show()
```

```python
x = [1,2,3]
y = [2,3,1]

plt.title('기본 bar 차트')
plt.bar(1,2)
plt.bar(2,3)
plt.bar(3,1)

# tick 설정
plt.xticks(x,['X1','X2','X3'])
plt.yticks(sorted(y),['one','two','three'])

# 각 축의 레이블
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

```python
x = [1,2,3]
y = [2,3,1]

plt.title('기본 bar 차트')
plt.barh(1,2)
plt.barh(2,3)
plt.barh(3,1)

# tick 설정
plt.xticks(x,['X1','X2','X3'])
plt.yticks(sorted(y),['one','two','three'])

# 각 축의 레이블
plt.xlabel('x축')
plt.ylabel('y축')

plt.show()
```

### 선그래프

```python
x = [10,20,30,40]
y = [1,4,9,16]

plt.title('Plot')
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(-np.pi,np.pi,256)
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

### 스템플롯

```python
x = np.linspace(-np.pi,np.pi,20)
y = np.cos(x)
plt.stem(x,y,'--',use_line_collection=True)
plt.show()
```

### pie 플롯

카테고리별로 값의 상대적인 비교를 할 때 ...

```python
data1 = [15,30,45,10]

plt.title('pie plot')
plt.pie(data1)
plt.show()
```

```python
data1 = [15,30,45,10]

plt.title('pie plot')

# 튀어 나오는 것 설정
option1 = [0, 0.1, 0, 0]
# 각 조각의 레이블
option2 = ['조각1','조각2','조각3','조각4']
# 각 조각별 색상
option3 = ['yellowgreen','gold','lightskyblue','lightcoral']

# explode : 튀어나오는 정도
# labels : 각 조각의 label
# colors : 색상
# autopct : 표시할 퍼센트 양식
# shadow : 그림자 표시 여부
# startangle : 시작 각도 값, 생략시 3시 방향
# textprops : 폰트 설정
# counterclock : 그래프를 그리는 방향. True(기본) - 시계 반대. False - 시계 방향
plt.pie(data1, explode=option1, labels=option2, colors=option3, autopct='%1.1f%%',
       shadow=True, startangle=90, textprops={'size':13}, counterclock=False)

plt.show()

plt.savefig('fig1.png', dpi=300)
```
