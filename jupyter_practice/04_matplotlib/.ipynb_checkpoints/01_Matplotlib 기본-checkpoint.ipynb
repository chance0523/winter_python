# matplotlib

파이썬에서 데이터 시각화를 위해 사용하는 라이브러리

https://matplotlib.org

**seaborn**과 더불어 가장 많이 사용하는 시각화 라이브러리

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 기본 플롯 그리기

```python
plt.title('Plot')
plt.plot([1,4,9,16])
plt.show()
```

```python
array1 = np.array([1,4,9,16])
plt.title('Plot')
plt.plot(array1)
plt.show()
```

### x축, y축 데이터 셋팅

```python
list1 = [10,20,30,40]
list2 = [1,4,9,16]

plt.title('Plot')
# 첫 번째 리스트가 x축, 두 번째 리스트가 y축에 셋팅된다.
plt.plot(list1, list2)
plt.show()
```

```python
list1 = [1,1,1,1]
list2 = [2,4,6,8]
list3 = [3,6,9,12]
list4 = [4,8,12,16]

list5 = [1,2,3,4]

plt.title('Plot')
# 첫 번째 리스트가 x축, 두 번째 리스트가 y축에 셋팅된다.
# x축의 리스트들의 첫번째 데이터를 모아 선 하나, 두번째 데이터를 모아 선 하나...
plt.plot([list1,list2,list3,list4],list5)
plt.show()
```

```python
list1 = [1,1,1,1]
list2 = [2,4,6,8]
list3 = [3,6,9,12]
list4 = [4,8,12,16]

list5 = [1,2,3,4]

plt.title('Plot')
# 첫 번째 리스트가 x축, 두 번째 리스트가 y축에 셋팅된다.
# 플롯의 수 만큼 그래프를 그린다.
plt.plot(list1,list5)
plt.plot(list2,list5)
plt.plot(list3,list5)
plt.plot(list4,list5)

plt.show()
```

```python
array1 = np.array([
    [1,1,1,1],
    [2,4,6,8],
    [3,6,9,12],
    [4,8,12,16]
])
array2 = np.array([1,2,3,4])

plt.title('Plot')
# 첫 번째 리스트가 x축, 두 번째 리스트가 y축에 셋팅된다.
# x축의 리스트들의 첫번째 데이터를 모아 선 하나, 두번째 데이터를 모아 선 하나...
plt.plot(array1, array2)
plt.show()
```

```python
array1 = np.array([
    [1,1,1,1],
    [2,4,6,8],
    [3,6,9,12],
    [4,8,12,16]
])
array2 = np.array([1,2,3,4])

plt.title('Plot')
# 첫 번째 리스트가 x축, 두 번째 리스트가 y축에 셋팅된다.
plt.plot(array1[0], array2)
plt.plot(array1[1], array2)
plt.plot(array1[2], array2)
plt.plot(array1[3], array2)

plt.show()
```

### 각 축의 레이블

```python
plt.title('Label')
plt.plot([10,20,30,40], [1,4,9,16])
plt.xlabel('X label')
plt.ylabel('Y label')
plt.show()
```

### 한글 사용

```python
plt.title('한글 타이틀')
plt.plot([10,20,30,40], [1,4,9,16])
plt.xlabel('X축 레이블')
plt.ylabel('Y축 레이블')
plt.show()
```

### 컴퓨터에 설치되어 있는 폰트 목록

### 기본 폰트 설정

```python
from matplotlib import rcParams
```

```python
plt.rc('font', family='Malgun Gothic')
plt.rc('font', size='16')

# 사용하는 폰트에 - 가 없을 경우 - 는 지정된 폰트가 아닌
# 원래의 기본 폰트를 사용하도록 지정한다.
plt.rcParams['axes.unicode_minus']=False
```

```python
plt.title('한글 타이틀')
plt.plot([10,20,30,40], [1,4,9,16])
plt.xlabel('X축 레이블')
plt.ylabel('Y축 레이블')
plt.show()
```

### 폰트 객체 직접 만들기

사용하고자 하는 폰트가 기본 폰트로 설정되지 않거나 특정 부분에서만 사용하고 싶을 때

```python
font_path = 'C:/Windows/Fonts/BMJUA_ttf.ttf'
font_obj = fm.FontProperties(fname=font_path, size=16)
```

```python
plt.title('한글 타이틀', fontproperties=font_obj)
plt.plot([10,20,30,40], [1,4,9,16])
plt.xlabel('X축 레이블', fontproperties=font_obj)
plt.ylabel('Y축 레이블', fontproperties=font_obj)
plt.show()
```

### 선 스타일

```python
plt.title('선 스타일(색상, 마커종류, 선 종류)')
plt.plot([10,20,30,40],[1,4,9,16],'r*--')
plt.plot([10,20,30,40],[16,9,4,1],'bs:')
plt.show()
```

### 상세 설정
<img src='image/4.png'/>

```python
plt.title('상세설정')
plt.plot([10,20,30,40], [1,4,9,16], c='b', lw=5, ls='--',
         marker='D', ms=15, mec='g', mew=5, mfc='r')
plt.show()
```

### 범위지정

```python
plt.title('범위지정')
plt.plot([10,20,30,40], [1,4,9,16], c='b', lw=5, ls='--',
         marker='D', ms=15, mec='g', mew=5, mfc='r')
plt.xlim(0,50)
plt.ylim(-10,30)
plt.show()
```

### 틱 설정

```python
plt.title('틱 설정')
plt.plot([10,20,30,40], [1,4,9,16])
plt.xticks([10,20,30,40,],['십','이십','삼십','사십'])
plt.yticks([1,4,9,16],['일','사','구','십육'])
plt.show()
```

### 그리드

```python
plt.title('틱 설정')
plt.plot([10,20,30,40], [1,4,9,16])
plt.grid(True)
plt.show()
```

### 범례
<img src='image/5.png'/>

```python
plt.title('범례')
plt.plot([10,20,30,40],[1,4,9,16],'r*--', label='빨강')
plt.plot([10,20,30,40],[16,9,4,1],'bs:', label='파랑')
plt.legend()    # 0(best)으로 세팅
# plt.legend(loc=10)
plt.show()
```
