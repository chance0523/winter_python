```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_csv(station_name):
    yearList = ['2016','2017','2018','2019','2020']
    for year in yearList: 
        if year != '2020':
            for month in range(1,13):
                df1 = pd.read_csv(f'{station_name}/{year}/{station_name}_{year}_{month}.csv')
        else:
            df1 = pd.read_csv(f'{station_name}/{year}/{station_name}_2020_1.csv')
```

```python
def read_csv_only_one(station_name, year, month):
    if year != '2020':
        df1 = pd.read_csv(f'{station_name}/{year}/{station_name}_{year}_{month}.csv')
    else:
        df1 = pd.read_csv(f'{station_name}/{year}/{station_name}_2020_1.csv')
    return df1
```

```python
# read_csv('Gangnam_Hakdongro')
df1 = read_csv_only_one('Gangnam_Hakdongro','2019','1')
```
