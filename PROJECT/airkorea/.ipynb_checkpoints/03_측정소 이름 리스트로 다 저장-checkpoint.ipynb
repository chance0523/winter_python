airkorea 사이트의 측정소 리스트 엑셀 파일에서 측정소 이름을 다 가져와 리스트로 만든다.

```python
import pandas as pd
```

```python
option1=[_ for _ in range(5)]
dosi_xls = pd.read_excel('D:/VSCode/winter_python/python/PROJECT/airkorea/dosi_measure_station.xls',names=option1)
doro_xls = pd.read_excel('D:/VSCode/winter_python/python/PROJECT/airkorea/doro_measure_station.xls',names=option1)
```

```python
dosi_xls.head()
```

```python
print(dosi_xls[2][3:])
print(dosi_xls[2][3:].count())
```

```python
def make_station_list(xls):
    station_list=[]
    cnt = xls[2][3:].count()
    for i in range(cnt):
        station_list.append(xls[2][i+3])
    return station_list
```

```python
dosi_station_list = make_station_list(dosi_xls)
doro_station_list = make_station_list(doro_xls)
```

```python
dosi = pd.DataFrame(dosi_station_list)
doro = pd.DataFrame(doro_station_list)
```

```python
dosi.to_csv('dosi_station_list.csv',index=False,header=False,encoding='utf-8-sig')
doro.to_csv('doro_station_list.csv',index=False,header=False,encoding='utf-8-sig')
```
